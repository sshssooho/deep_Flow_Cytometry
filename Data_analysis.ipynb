```python
import os
import fcsparser
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
os.listdir('../../data/')
```

```python
file_dict = {}
for files in os.listdir('../../data/'):
    patient_num = files.split('[')[-3][:-1]
    file_type = files.split('[')[-1][:-5]
    if patient_num in file_dict:
        file_dict[patient_num].append(file_type)
    else:
        file_dict[patient_num] = [file_type]
```

```python
flag = True
for key, value in file_dict.items():
    if 'WNR' not in value:
        print(key)
        flag = False
flag
```

```python
from mpl_toolkits.mplot3d import Axes3D

sns.set(style = "darkgrid")


fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection = '3d')

x = data['Side Fluorescence Signal']
y = data['Forward Scatter Signal']
z = data['Side Scatter Signal']

ax.set_xlabel('Side Fluorescence Signal')
ax.set_ylabel('Forward Scatter Signal')
ax.set_zlabel('Side Scatter Signal')

ax.scatter(x, y, z)		

plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2)
a = kmeans.fit(data)
labels = kmeans.labels_
colors = {0:'red', 1:'green', 2:'blue'}

def f(x):
    return colors[x]

color_labels = list(map(f, labels))

sns.set(style = "darkgrid")
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection = '3d')

x = data['Side Fluorescence Signal']
y = data['Forward Scatter Signal']
z = data['Side Scatter Signal']

ax.set_xlabel('Side Fluorescence Signal')
ax.set_ylabel('Forward Scatter Signal')
ax.set_zlabel('Side Scatter Signal')

ax.scatter(x, y, z, c=color_labels)		

plt.show()
```
